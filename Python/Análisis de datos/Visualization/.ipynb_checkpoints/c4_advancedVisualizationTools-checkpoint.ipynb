```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # needed for waffle Charts
```

## DATAFRAME

**Read DataFrame**

```python
## Paths of the file of data analysis
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx"
path = "C:\\Users\\USUARIO1\\Desktop\\MyPython\\Análisis de datos\\Visualization\\Canada.xlsx"

## Read the file either through url or direction file
df = pd.read_excel(url, sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)

df.head()
```

**Make changes to de data set**

```python
## Remove columns that we don't consider important
df.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
## Rename columns whit a more specific name
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
## Use countries as name of columns
df.set_index('Country', inplace=True)
## Convert all the values of the columns to string
df.columns = list(map(str, df.columns))
## Add column whit the sum of migrations values
df['Total'] = df.sum(axis = 1)

years = list(map(str, range(1980, 2014)))

df.head()
```

## WAFFLE CHARTS

```python
# let's create a new dataframe for these three countries 
df_dsn = df.loc[['Denmark', 'Norway', 'Sweden'], :]

# let's take a look at our dataframe
df_dsn
```

```python
percentages = [((df_dsn.loc[country, 'Total'])/(sum(df_dsn['Total']))) for country in df_dsn.index]

for (i,  percentage) in list(enumerate(percentages)):
    print("The percentage of ", df_dsn.index.values[i], " is: ", percentage)
#rof
```

```python
## Sizes of chart
width = 40 # width of chart
height = 10 # height of chart
total_num_tiles = width * height # total number of tiles

tiles_per_category = [round(p * total_num_tiles) for p in percentages]

for (i,  percentage) in list(enumerate(tiles_per_category)):
    print("Number of tales ", df_dsn.index.values[i], " is: ", percentage)
#rof
```

```python
waffle_chart = np.zeros((height, width))

cont = 1
j = 0
for col in range(width):
    for row in range(height):
        
        if ( cont > sum(tiles_per_category[0:j])):
            j = j + 1
        #fi
        
        waffle_chart[row, col] = j
        cont = cont + 1
    #rof
#rof

print(np.array(waffle_chart))
```

```python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
```

```python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])


# create legend
legend_handles = []
for i, category in enumerate(df_dsn.index.values):
    label_str = category + ' (' + str(df_dsn['Total'][i]) + ')'
    color_val = colormap(i*0.5)
    legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

# add legend to chart
plt.legend(handles=legend_handles,
           loc='lower center',                 ## position
           ncol=len(df_dsn.index.values),      ## large box
           bbox_to_anchor=(0., -0.2, 0.95, .1) ## final positio
          )
```

## Word Clouds

```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

path_alice_novel = "C:\\Users\\USUARIO1\\Desktop\\MyPython\\Análisis de datos\\Visualization\\alice_novel.txt"

# open the file and read it into a variable alice_novel
alice_novel = open(path_alice_novel, 'r').read()
```

```python
## words that will be used in the plot
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
# instantiate a word cloud object
alice_wc = WordCloud(font_path=None, width = 1500, height=500,  
            max_words=200,  stopwords=None, background_color='whitesmoke', max_font_size=None, font_step=1, mode='RGB', 
            collocations=True, colormap=None, normalize_plurals=True)

# generate the word cloud
alice_wc.generate(alice_novel)

# display the word cloud
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

**Add word from the cloud. Change height and width**

```python
# add the words said to stopwords
stopwords.add('said')

# re-generate the word cloud
alice_wc.generate(alice_novel)

# display the cloud
fig = plt.figure()
fig.set_figwidth(15) # set width
fig.set_figheight(20) # set height
interpolation  = ['antialiased', 'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos', 'blackman]'

plt.imshow(alice_wc, interpolation="bilinear", aspect="equal")
plt.axis('off')
plt.show()
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets


interpolation_options  = ['antialiased', 'none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos', 'blackman']

# add the words said to stopwords
stopwords.add('said')

# re-generate the word cloud
alice_wc.generate(alice_novel)

def change_interpolation(inter):
    # display the cloud
    fig = plt.figure()
    fig.set_figwidth(15) # set width
    fig.set_figheight(20) # set height


    plt.imshow(alice_wc, interpolation=inter, aspect="equal")
    plt.axis('off')
    plt.show()
                  
inter = widgets.Dropdown(options=interpolation_options, value='bilinear', description='Interpolation:', disabled=False)

widgets.interactive(change_interpolation, inter = inter)
```

**Cloud words whit DataFrame**

```python
max_words = 90
word_string = ''
for country in df.index.values:
    # check if country's name is a single-word name
    # enter the countries that percentage * 90 is more than 1
    if len(country.split(' ')) == 1:
        repeat_num_times = int(df.loc[country, 'Total']/df['Total'].sum()*max_words)
        word_string = word_string + ((country + ' ') * repeat_num_times)
                                     
# display the generated text
word_string
```

```python
# create the word cloud
wordcloud = WordCloud(font_path=None, width = 1500, height=500,  
            max_words=2000,  stopwords=None, background_color='whitesmoke', max_font_size=None, font_step=1, mode='RGB', 
            collocations=True, colormap=None, normalize_plurals=True).generate(word_string)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(wordcloud, interpolation="bilinear", aspect="equal")
plt.axis('off')
plt.show()
```

## REGRESSION PLOTS

```python
# we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df[years].sum(axis=0))

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace=True)

# rename columns
df_tot.columns = ['year', 'total']

# change the years to type float (useful for regression later on)
df_tot['year'] = df_tot['year'].astype(float)

# view the final dataframe
df_tot.head()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 7))
## Configurate atributtes of plot. Maker: form of point {+,*,<,o}. scatter_ kws: size of the marker
ax = sns.regplot(x='year', y='total', data=df_tot, color='red', marker='.', scatter_kws={'s': 200})

ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-label
ax.set_title('Total Immigration to Canada from 1980 - 2013') # add title
```
